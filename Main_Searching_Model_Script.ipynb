```python
import os
import requests
from io import BytesIO
from PIL import Image
import numpy as np
import keras
from keras.applications.resnet50 import ResNet50, preprocess_input
import pickle
import ipywidgets as widgets
from IPython.display import display, clear_output

# Load the KNN model and data

#  ---------------------------------------------------------------------------------------------------------------------------------------------------------------------

      #  Copy path via right click from your mounted drive & paste here , at when cell running first time it may take little longer
with open('/content/drive/MyDrive/Models/prediction_by_knn_model.pkl', 'rb') as f:
    knn = pickle.load(f)


       #  Copy path via right click from your mounted drive & paste here , at when cell running first time it may take little longer
with open('/content/drive/MyDrive/Models/product_maped_to_features', 'rb') as f:
    data = pickle.load(f)


    # -------------------------------------------------------------------------------------------------------------------------------------------------------------------------

links = [data['links'][i] for i, f in enumerate(data['feature_maps']) if f is not None]

# Load ResNet50 model without the top layers
base_model = ResNet50(weights='imagenet', include_top=False)

# Function to preprocess and get feature maps
def get_feature_map(image_path):
    if os.path.exists(image_path):  # Check if the path is a local file path
        img = Image.open(image_path).resize((224, 224))
    else:  # Otherwise, assume it's a URL
        response = requests.get(image_path)
        img = Image.open(BytesIO(response.content)).resize((224, 224))

    x = keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature_map = base_model.predict(x)
    return feature_map.flatten()

# Function to find the most similar product link
def find_similar_product(image_path):
    feature_map = get_feature_map(image_path)
    distances, indices = knn.kneighbors([feature_map])
    return links[indices[0][0]]

# Function to handle image upload and display input image and output link
def on_image_upload(change):
    # Clear previous output
    clear_output()

    # Handle image upload
    image_data = change['new']
    file_info = next(iter(image_data.values()))
    image_bytes = file_info['content']
    image = Image.open(BytesIO(image_bytes))
    image_np = np.array(image)

    # Save the uploaded image to a file
    uploaded_image_path = '/content/uploaded_image.png'
    image.save(uploaded_image_path)

    # Display the uploaded image with heading
    display(widgets.HTML("<h3>Input Image</h3>"))
    display(image)

    # Find and display the most similar product link
    similar_product_link = find_similar_product(uploaded_image_path)
    display(widgets.HTML(f"<h3>Output</h3><p>The most similar product link: <a href='{similar_product_link}' target='_blank'>{similar_product_link}</a></p>"))

# Create an image upload widget with custom description, size, and border
upload_widget = widgets.FileUpload(
    accept='image/*',  # Accept only images
    multiple=False,  # Single file upload
    description="Click upload & wait for 40 seconds",  # Custom description
    layout=widgets.Layout(width='300px', height='50px', border='solid 2px green'),  # Size and border
    style={'font_weight': 'bold'}  # Bold text
)

# Link the widget to the handler function
upload_widget.observe(on_image_upload, names='value')

# Display the upload widget
display(upload_widget)
```

```python
import os
import cv2
from PIL import Image
from ultralytics import YOLO
# must do pip install ultralytics
import ipywidgets as widgets
from IPython.display import display, clear_output
import io
import numpy as np
import keras
from keras.applications.resnet50 import ResNet50, preprocess_input
import pickle
import requests
from io import BytesIO

# Function to create output directory
def create_output_dir(output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

# Function to save and return image cutouts
def get_cutouts(image, boxes):
    cutouts = []
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)
        cutout = image[y1:y2, x1:x2]
        cutouts.append(cutout)
    return cutouts

# Function to handle image upload and display cutouts
def on_image_upload(change):
    # Clear previous output
    clear_output()

    # Handle image upload
    image_data = change['new']
    file_info = next(iter(image_data.values()))
    image_bytes = file_info['content']
    image = Image.open(io.BytesIO(image_bytes))
    image_np = np.array(image)

    # Display the uploaded image with heading
    display(widgets.HTML("<div><h2>Input Image</h2> <h3>at below use select button to select the product you want (IT MAY TAKE SOME TIME )<h3></div>"))
    display(image)

    # Load the YOLOv8 model
    model = YOLO('yolov8n.pt')  # Use the appropriate model version

    # Convert image to RGB for YOLO
    image_rgb = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)

    # Perform object detection
    results = model.predict(source=image_rgb)

    # Extract bounding boxes
    boxes = results[0].boxes.xyxy.cpu().numpy()  # or use .xywh for (x, y, w, h)

    # Get cutouts
    cutouts = get_cutouts(image_np, boxes)

    # Display cutouts with selection buttons
    display_cutouts_with_buttons(cutouts)

# Function to display cutouts with selection buttons
def display_cutouts_with_buttons(cutouts):
    cutout_widgets = []
    for i, cutout in enumerate(cutouts):
        cutout_image = Image.fromarray(cutout)
        cutout_image.thumbnail((200, 200))
        cutout_widget = widgets.Image(value=convert_image_to_bytes(cutout_image), format='png' ,layout=widgets.Layout(width='300px', height='300px', border='solid 4px green') )
        button = widgets.Button(description=f'Select {i}',  layout=widgets.Layout(width='auto', height='50px', border='solid 4px green'),  style={'font_weight': 'bold' , 'font_size' : '40px'})
        button.on_click(lambda b, cutout=cutout: on_select_cutout(cutout))
        cutout_widgets.append(widgets.VBox([cutout_widget, button]))
    display(widgets.HBox(cutout_widgets , display='flex'))

# Function to convert image to bytes
def convert_image_to_bytes(image):
    with io.BytesIO() as output:
        image.save(output, format="PNG")
        return output.getvalue()

# Function to handle cutout selection
def on_select_cutout(cutout):
    # Clear previous output
    clear_output()
    # Display the selected cutout
    cutout_image = Image.fromarray(cutout)
    display(widgets.HTML("<div><h3>Selected Cutout </h3><p> wait 40 seconds to get output link</p></div>"))
    display(cutout_image)

    # Save the selected cutout to a file
    cutout_path = '/content/selected_cutout.png'
    cutout_image.save(cutout_path)

    # Find and display the most similar product link
    similar_product_link = find_similar_product(cutout_path)
    display(widgets.HTML(f"<div><h3>Output</h3><h4>The most similar product link: <a href='{similar_product_link}' target='_blank'>{similar_product_link}</a></h4><p>NOW RUN THE CELL AGAIN TO GIVE ANOTHER TEST IMAGE</p></div>"))

# Function to preprocess and get feature maps
def get_feature_map(image_path):
    if os.path.exists(image_path):  # Check if the path is a local file path
        img = Image.open(image_path).resize((224, 224))
    else:  # Otherwise, assume it's a URL
        response = requests.get(image_path)
        img = Image.open(BytesIO(response.content)).resize((224, 224))

    x = keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature_map = base_model.predict(x)
    return feature_map.flatten()

#  ---------------------------------------------------------------------------------------------------------------------------------------------------------------------

      #  Copy path via right click from your mounted drive & paste here , at when cell running first time it may take little longer
with open('/content/drive/MyDrive/Models/prediction_by_knn_model.pkl', 'rb') as f:
    knn = pickle.load(f)


       #  Copy path via right click from your mounted drive & paste here , at when cell running first time it may take little longer
with open('/content/drive/MyDrive/Models/product_maped_to_features', 'rb') as f:
    data = pickle.load(f)


    # -------------------------------------------------------------------------------------------------------------------------------------------------------------------------


links = [data['links'][i] for i, f in enumerate(data['feature_maps']) if f is not None]

# Load ResNet50 model without the top layers
base_model = ResNet50(weights='imagenet', include_top=False)

# Function to find the most similar product link
def find_similar_product(image_path):
    feature_map = get_feature_map(image_path)
    distances, indices = knn.kneighbors([feature_map])
    return links[indices[0][0]]

# Create an image upload widget
upload_widget = widgets.FileUpload(
    description = 'click here to upload & wait for 50 seconds',
    accept='image/*',  # Accept only images
    multiple=False,  # Single file upload
    layout=widgets.Layout(width='400px', height='50px', border='solid 4px green'),  # Adjust the size and border
    style={'font_weight': 'bold' , 'font_size' : '40px'}  # Make the text bold
)

# Link the widget to the handler function
upload_widget.observe(on_image_upload, names='value')

# Display the upload widget
display(upload_widget)
```

# **FOLLOWING CODES ARE USED TO TRAIN MODELS**
**CODE TO MAP IMG FEATURES WITH PRODUCT LINK VIA transfer learning in RESNET50** ( ***NO NEED TO RUN THIS CELL*** )

```python
import keras
from keras.applications.resnet50 import ResNet50, preprocess_input
import numpy as np
import pandas as pd
import requests
from io import BytesIO
from PIL import Image
import pickle

# Load ResNet50 model without the top layers
base_model = ResNet50(weights='imagenet', include_top=False)

# Function to preprocess and get feature maps
def get_feature_map(img_url):
    response = requests.get(img_url)
    img = Image.open(BytesIO(response.content)).resize((224, 224))
    x = keras.utils.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature_map = base_model.predict(x)
    return feature_map.flatten()

# Load the CSV file


# -----------------------------------------------------------------------------------------------------------
  # PASTE THE CSV FILE OF DATA ON WHICH YOU WANT TO TRAIN MODEL
csv_file_path = '/content/all category.csv'
df = pd.read_csv(csv_file_path)
# --------------------------------------------------------------------------------------------------------



# Process each image and get its feature map
feature_maps = []

for index, row in df.iterrows():
    img_url = row['imgurl']
    try:
        feature_map = get_feature_map(img_url)
        feature_maps.append(feature_map)
    except Exception as e:
        print(f"Error processing image at {img_url}: {e}")
        feature_maps.append(None)  # Append None if there's an error

# Save the feature maps and links to a file
data = {
    'links': df['linkn'].tolist(),
    'feature_maps': feature_maps
}

with open('product_maped_to_features', 'wb') as f:
    pickle.dump(data, f)

print("Feature maps have been saved to product_maped_to_features.pkl")
```
